```python
import numpy  as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'A': [1, 20, 4,0],
    'B': [4, 5, 6,0]
})
```

```python
df
```

```python
df.sort_values(by="A") # ascending sort
```

```python
df.sort_values(by="A", ignore_index=True) # ascending sort
```

```python
df.sort_values(by="A", ignore_index=True,
              ascending=False) # descending sort

# add inplace = True to save this operation!
```
