```python
# Merge on index!
```

```python
df1=pd.DataFrame(
    {"letter":["a","b","c"],
     "num":range(3)})
df2=pd.DataFrame(
    {"value":[3,5,7]},
    index=["a","b","e"])
```

```python
df1
```

```python
df2
```

```python
# df1 -> letter column
# df2 -> index
```

```python
pd.merge(df1,df2,left_on = 'letter', right_index=True, how = 'inner')
```

```python
import pandas as pd

employees = pd.DataFrame({
    "emp_id": [101, 102, 103, 104],
    "name": ["Amit", "Neha", "Ravi", "Sara"],
    "dept_id": [1, 2, 2, 4]
})

departments = pd.DataFrame({
    "dept_id": [1, 2, 3],
    "dept_name": ["HR", "Engineering", "Sales"]
})
```

```python
# 1. Write code to perform an INNER JOIN between employees and departments on dept_id.
pd.merge(employees, departments, on="dept_id", how="inner")
```

```python
# 2. Write code to perform a LEFT JOIN that keeps all employees, even if they donâ€™t belong to a department.
pd.merge(employees, departments, on="dept_id", how="left")
```

```python
# 3. Write code to perform a RIGHT JOIN that keeps all departments, even if no employee belongs to them.
pd.merge(employees, departments, on="dept_id", how="right")
```

```python
# 4. Write code to perform an OUTER JOIN that keeps all employees and all departments.
pd.merge(employees, departments, on="dept_id", how="outer")
```

```python
# 5. Write code to identify employees whose department
# information is missing after the join.
joined = pd.merge(employees, departments, on="dept_id", how="left")
joined[joined["dept_name"].isna()]
```
