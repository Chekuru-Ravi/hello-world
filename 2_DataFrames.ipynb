```python
import pandas as pd
df = pd.DataFrame({ "gender": ["M", "F", "F", "M", "M", "F"], "city": ["Delhi", "Delhi", "Mumbai", "Mumbai", "Delhi", "Mumbai"], "wt": [70, 55, 60, 80, 75, 65], "age": [25, 23, 27, 30, 28, 26] })
```

```python
df
```

```python
# 1. Using groupby('gender'), calculate the mean weight for each gender.
# 2. Using groupby('city'), calculate the maximum age in each city.
# 3. Using groupby(['gender', 'city']), calculate the sum of weight for each group.
# 4. Using groupby('gender'), calculate the mean of both weight and age.
# 5. Groups the data by gender : # Calculates the average (mean) and maximum of the wt column # Calculates the minimum of the age column
```

```python
df.groupby("gender")[['wt', 'age']].mean()
```

```python
df.groupby("city")[['age']].max()
```

```python
df.groupby(['gender', 'city'])[['wt']].sum()
```

```python
df.groupby('gender')[['wt', 'age']].agg(['mean'])
```

```python
df.groupby('gender')[['wt', 'age']].agg({'wt': ['mean', 'max'], 'age': ['min']})
#
```

```python
import pandas as pd
employees = pd.DataFrame({ "emp_id": [101, 102, 103, 104], "name": ["Amit", "Neha", "Ravi", "Sara"], "dept_id": [1, 2, 2, 4] })
departments = pd.DataFrame({ "dept_id": [1, 2, 3], "dept_name": ["HR", "Engineering", "Sales"] })
```

```python
# 1. Write code to perform an INNER JOIN between employees and departments on dept_id.
# 2. Write code to perform a LEFT JOIN that keeps all employees, even if they dont belong to a department.
# 3. Write code to perform a RIGHT JOIN that keeps all departments, even if no employee belongs to them.
# 4. Write code to perform an OUTER JOIN that keeps all employees and all departments.
# 5. Write code to identify employees whose department information is missing after the join.
```

```python
pd.merge(employees, departments, on='dept_id', how='inner')
```

```python
pd.merge(employees, departments, on='dept_id', how='left')
```

```python
pd.merge(employees, departments, on='dept_id', how='right')
```

```python
pd.merge(employees, departments, on='dept_id', how='outer')
```

```python
merged_df = pd.merge(employees, departments, on='dept_id', how='outer')
merged_df[merged_df['dept_name'].isna()]
```
